# NER Training
## Train CPU-based ner (on spacy lg model)


```python
from spacy.cli.train import train
import os.path
BASE_DIR = "../data/v4"
use_gpu = 0
```

```python
root = BASE_DIR.rstrip('/')
suffix = 'cpu' if use_gpu == -1 else 'gpu'
train("../configs/gpu_config.cfg", root + "-model." + suffix, overrides={"paths.train": os.path.join(root, "train.spacy"), "paths.dev": os.path.join(root, "test.spacy")}, use_gpu=use_gpu)
```
